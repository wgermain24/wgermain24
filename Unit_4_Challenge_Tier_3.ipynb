**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    """Return the ratio of the average price of 2018 to the average price of 1998"""
    y1998 = float(d['Average_price'][d['Year'] == 1998]) # Save the average price for 1998
    y2018 = float(d['Average_price'][d['Year'] == 2018]) # Save the average price for 2018
    ratio =[y2018/y1998]
    return ratio
```

```python
# We test if the function works like we define it
print(create_price_ratio(df_year[df_year['London_Borough'] == 'LONDON']))
```

```python
final ={}
```

```python
# Using of a loop For to go through the DataFrame df_year, compute the ratio y2018/y1998 and stock them in the dictionary

for b in df_year['London_Borough'].unique():
    borough = df_year[df_year['London_Borough'] == b] # creation of borough as variable to simplify the code
    final[b] = create_price_ratio(borough)    # add to the dictionary, the Boroughs as key and the ratios as values
print(final)
```

```python
df_ratios = pd.DataFrame(final) # conversion ratio dictionary in DataFrame
```

```python
df_ratios.head()
```

```python
df_ratios_T = df_ratios.T   # transpose the new DataFrme df_ratios
df_ratios.head()
```

```python
df_ratios = df_ratios_T.reset_index() 
```

```python
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'ratio_2018_1998'}, inplace = True) # Rename the columns : index = 'Borough' and 0 = 'ratio_2018_1998'
df_ratios.head()
```

```python
top_15 = df_ratios.sort_values(by='ratio_2018_1998', ascending = False).head(15)
```
